<h4> Extract the year out of the date, and set as an index</h4>

```python
df4 = df
df4['year'] = df4.date
df4['year'] = df4['year'].str.slice(0, 4)
df4 = df4.set_index('year')

df4
```

<h2> Use Group by to Obtain Subsets of the Data </h2>
    <h3> Plot The country/year for total vaccinations per hundred population</h3>

```python

df4.groupby(['country','year'])['total_vaccinations_per_hundred'].mean().plot( legend=True, rot=75, figsize=(15,10))
```

 <h3> Plot number of people fully vaccinated by vaccine types</h3>

```python
vaccines = df.groupby('vaccines')['people_fully_vaccinated'].sum()
plt = vaccines[['Oxford/AstraZeneca','Johnson&Johnson','Pfizer/BioNTech','Moderna']].plot.bar(legend=True, figsize=(15,10), rot=75)

```

<h3> Visualize total vaccinations per hundred by country. </h3>

```python
c=df.groupby(['year','country'])['total_vaccinations_per_hundred'].mean()
c.groupby(['country']).plot( legend=True, rot=75, figsize=(15,10))

```
