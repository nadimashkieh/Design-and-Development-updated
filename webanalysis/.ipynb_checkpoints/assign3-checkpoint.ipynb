```python
#import libraries 
import urllib.request, urllib.error, urllib.parse
#assign a variable for the needed url 
url = 'https://www.cia.gov/the-world-factbook/countries/lebanon/'
#assign a variable to store the url content; read the contents
response = urllib.request.urlopen(url)
webContent = response.read()

print(webContent[0:1000])
```

```python
#get contents without the html, assign a new variable as a str, identify start and end location
#assign the new content to the str variable
CleanContent = str(webContent)
startLoc = CleanContent.find("Introduction")
endLoc = CleanContent.find("in 2019")

print(startLoc)
print(endLoc)

CleanContent = CleanContent[startLoc: endLoc]
print(CleanContent[0: 500])
```

```python
# assign a variable to value 0, another to an empty string, create a for loop to check for characters
# that the content does not have the html part

inside = 0
text = ''

for char in CleanContent:
    if char == '<':
        inside = 1
    elif (inside == 1 and char == '>'):
        inside = 0
    elif inside== 1:
        continue
    else: 
        text += char

print(text[0: 5000])
```

```python
f = open('../week3/lebdata.txt','w')
f.write(text)
f.close()
```
